```python
#import library
import nltk
from nltk import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import pandas as pd
```

```python
#set stopwords language
stop_words = stopwords.words('english')
stemmer = PorterStemmer()
```

```python
#read data from csv
df = pd.read_csv('save.data_film.csv', sep=",")
data = list(df["description"].astype(str).str.lower())
word_processing_cleaned = []
word_processing_stemmed = []
for d in data:
    token = word_tokenize(d)
    cleaned_token = ""
    for word in token:
        if word not in stop_words:
            cleaned_token += word +" "
    print('Cleaned Token Before = ',d)
    print('Cleaned Token After = ',cleaned_token)
    word_processing_cleaned.append(cleaned_token)
    token_stemming = word_tokenize(cleaned_token)
    word_stemmed = ""
    for ws in token_stemming:
        word_stemmed += stemmer.stem(ws) + " "
    print('Cleaned Token After Stem = ',word_stemmed)
    word_processing_stemmed.append(word_stemmed)
```

```python
# save to csv
dict_cleaned = {'word':word_processing_cleaned}
df = pd.DataFrame(dict_cleaned) 
#save cleaned word token 
df.to_csv('post_processing_data_film_cleaned.csv') 

df
```

```python
dict_stemmed = {'word':word_processing_stemmed}
df = pd.DataFrame(dict_stemmed) 
#save stemmed word token to csv
df.to_csv('post_processing_data_film_stemmed.csv') 

df
```
